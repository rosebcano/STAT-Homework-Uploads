```python
import matplotlib.pyplot as plt

x_values = [0, 1, 2, 3, 4, 5]
y_values = [2, 4, 6, 14, 10, 12]

plt.plot(x_values, y_values, 'r--', linewidth=2, markersize=10, markerfacecolor='green', markeredgecolor='green', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Dashed Line with Markers')
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = [0, 1, 2, 3, 4, 5]
y_values_1 = [2, 4, 6, 8, 10, 12]
y_values_2 = [5, 10, 15, 20, 25, 30]
y_values_3 = [4, 6, 8, 10, 12, 14]

plt.plot(x_values, y_values_1, 'r', label='Line 1')
plt.plot(x_values, y_values_2, 'g', label='Line 2')
plt.plot(x_values, y_values_3, 'b', label='Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}

# Bar chart
plt.bar(marks.keys(), marks.values())
plt.title('Student Grades')
plt.xlabel('Student Name')
plt.ylabel('Grade')
plt.show()

# Pie chart
plt.pie(marks.values(), labels=marks.keys(), autopct='%1.1f%%')
plt.title('Student Grades')
plt.legend()
plt.show()
```
