```r
library(ggplot2)

df <- data.frame(
  name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
  salary = c(623.30, 515.20, 611.00, 729.00, 843.25),
  stringsAsFactors = FALSE
)

highest_salary <- max(df$salary)
median_salary <- median(df$salary)
lowest_salary <- min(df$salary)

pie_df <- data.frame(
  salary = c(highest_salary, lowest_salary, median_salary),
  label = c("Highest Salary", "Lowest Salary", "Median Salary")
)

ggplot(pie_df, aes(x="", y=salary, fill=label)) + 
  geom_bar(stat="identity", width=1, color="white") +
  coord_polar("y", start=0) + 
  labs(title="Salary Summary", fill="") +
  theme_void() +
  theme(legend.position="bottom")
```
